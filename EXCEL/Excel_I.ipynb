```python
## Change the excel

import openpyxl
from openpyxl import Workbook

wb = Workbook()

ws = wb.create_sheet("Mysheet1")
ws = wb.create_sheet("Mysheet2")

ws['A1'] = 42
ws.append([1, 2, 3])
wb.save("sample.xlsx")
```

## Load and iterate over notebook

```python
from openpyxl import load_workbook
wb2 = load_workbook('Excel I exercise.xlsx')

wb2.sheetnames
regions = wb2['REGIONS']
```

```python
for row in regions.iter_rows(min_row=1, max_row=4, min_col=1, max_col=4):
    print('changing row...')
    for cell in row:
        print(cell.value)
```
